### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Suma todos los valores de la columna 'precio'
total_precios = tienda['Precio'].sum()
total_precios2 = tienda2['Precio'].sum()
total_precios3 = tienda3['Precio'].sum()
total_precios4 = tienda4['Precio'].sum()

# Muestra el resultado
print(f"La suma total de todos los precios de los productos de la tienda 1 es: {total_precios}")
print(f"La suma total de todos los precios de los productos de la tienda 2 es: {total_precios2}")
print(f"La suma total de todos los precios de los productos de la tienda 3 es: {total_precios3}")
print(f"La suma total de todos los precios de los productos de la tienda 4 es: {total_precios4}")
```

```python
# Muestra una lista de todos los nombres de las columnas en el DataFrame
print(tienda.columns)
```

# 2. Ventas por categoría

```python

ventas_por_categoria = tienda.groupby('Categoría del Producto')['Precio'].sum()
ventas_por_categoria2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
ventas_por_categoria3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
ventas_por_categoria4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()

# Muestra el resultado
print("Ventas totales por categoría de todas las Tiendas:")
print("TIENDA 1")
print(ventas_por_categoria)
print("------------------------------------")
print("TIENDA 2")
print(ventas_por_categoria2)
print("------------------------------------")
print("TIENDA 3")
print(ventas_por_categoria3)
print("------------------------------------")
print("TIENDA 4")
print(ventas_por_categoria4)
print("------------------------------------")
```

```python
# Encuentra el nombre de la categoría con la venta más alta
categoria_mas_vendida = ventas_por_categoria.idxmax()
categoria_mas_vendida2 = ventas_por_categoria2.idxmax()
categoria_mas_vendida3 = ventas_por_categoria3.idxmax()
categoria_mas_vendida4 = ventas_por_categoria4.idxmax()


# Muestra el resultado
print(f"\nLa categoría que más vende en Tienda 1 es: {categoria_mas_vendida}")
print(f"\nLa categoría que más vende en Tienda 2 es: {categoria_mas_vendida2}")
print(f"\nLa categoría que más vende en Tienda 3 es: {categoria_mas_vendida3}")
print(f"\nLa categoría que más vende en Tienda 4 es: {categoria_mas_vendida4}")
```

# 3. Calificación promedio de la tienda


```python
promedio_satisfaccion = tienda['Calificación'].mean()
promedio_satisfaccion2 = tienda2['Calificación'].mean()
promedio_satisfaccion3 = tienda3['Calificación'].mean()
promedio_satisfaccion4 = tienda4['Calificación'].mean()

# Muestra el resultado
print(f"El grado de satisfacción promedio de los clientes en la tienda 1 es: {promedio_satisfaccion:.2f}")
print(f"El grado de satisfacción promedio de los clientes en la tienda 2 es: {promedio_satisfaccion2:.2f}")
print(f"El grado de satisfacción promedio de los clientes en la tienda 3 es: {promedio_satisfaccion3:.2f}")
print(f"El grado de satisfacción promedio de los clientes en la tienda 4 es: {promedio_satisfaccion4:.2f}")

# Redondea el promedio al entero más cercano
promedio_redondeado = round(promedio_satisfaccion)
promedio_redondeado2 = round(promedio_satisfaccion2)
promedio_redondeado3 = round(promedio_satisfaccion3)
promedio_redondeado4 = round(promedio_satisfaccion4)

# Muestra el resultado
print(f"El grado de satisfacción promedio redondeado en la tienda 1 es: {promedio_redondeado}")
print(f"El grado de satisfacción promedio redondeado en la tienda 2 es: {promedio_redondeado2}")
print(f"El grado de satisfacción promedio redondeado en la tienda 3 es: {promedio_redondeado3}")
print(f"El grado de satisfacción promedio redondeado en la tienda 4 es: {promedio_redondeado4}")
```

# 4. Productos más y menos vendidos

```python
# Agrupa los productos y suma sus precios para obtener el total de ventas por producto
ventas_por_producto = tienda.groupby('Producto')['Precio'].sum()
ventas_por_producto2 = tienda2.groupby('Producto')['Precio'].sum()
ventas_por_producto3 = tienda3.groupby('Producto')['Precio'].sum()
ventas_por_producto4 = tienda4.groupby('Producto')['Precio'].sum()


print("Ventas totales (por monto) por cada producto:")
print("TIENDA 1")
print(ventas_por_producto)
print("------------------------------------")
print("TIENDA 2")
print(ventas_por_producto2)
print("------------------------------------")
print("TIENDA 3")
print(ventas_por_producto3)
print("------------------------------------")
print("TIENDA 4")
print(ventas_por_producto4)
print("------------------------------------")
# Encuentra el nombre del producto que generó más ingresos
producto_mas_vendido = ventas_por_producto.idxmax()
producto_mas_vendido2 = ventas_por_producto2.idxmax()
producto_mas_vendido3 = ventas_por_producto3.idxmax()
producto_mas_vendido4 = ventas_por_producto4.idxmax()

# Encuentra el nombre del producto que generó menos ingresos
producto_menos_vendido = ventas_por_producto.idxmin()
producto_menos_vendido2 = ventas_por_producto2.idxmin()
producto_menos_vendido3 = ventas_por_producto3.idxmin()
producto_menos_vendido4 = ventas_por_producto4.idxmin()

# Muestra los resultados
print(f"\nEl producto que más vende en Tienda 1 es: {producto_mas_vendido}")
print(f"El producto que menos vende en Tienda 1 es: {producto_menos_vendido}")

print(f"\nEl producto que más vende en Tienda 2 es: {producto_mas_vendido2}")
print(f"El producto que menos vende en Tienda 2 es: {producto_menos_vendido2}")

print(f"\nEl producto que más vende en Tienda 3 es: {producto_mas_vendido3}")
print(f"El producto que menos vende en Tienda 3 es: {producto_menos_vendido3}")

print(f"\nEl producto que más vende en Tienda 4 es: {producto_mas_vendido4}")
print(f"El producto que menos vende en Tienda 4 es: {producto_menos_vendido4}")
```

# 5. Envío promedio por tienda

```python
promedio_envio = tienda['Costo de envío'].mean()
promedio_envio2 = tienda2['Costo de envío'].mean()
promedio_envio3 = tienda3['Costo de envío'].mean()
promedio_envio4 = tienda4['Costo de envío'].mean()

# Muestra el resultado. El :.2f es para mostrar solo dos decimales.
print(f"El costo de envío promedio de la tienda 1 es: {promedio_envio:.2f}")
print(f"El costo de envío promedio de la tienda 2 es: {promedio_envio2:.2f}")
print(f"El costo de envío promedio de la tienda 3 es: {promedio_envio3:.2f}")
print(f"El costo de envío promedio de la tienda 4 es: {promedio_envio4:.2f}")
```
